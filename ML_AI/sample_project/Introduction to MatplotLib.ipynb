## Matplot Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots()

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and Show (You save the whole figure)
fig.savefig("images/sample-plot.png")
plt.show()
```

```python
# make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title = "Nut Butter Store", ylabel="Price ($)", xlabel="Butters");
plt.show();
```

```python
# Horizontal bar chart
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
ax.set(title = "Nut Butter Store", xlabel="Price ($)", ylabel="Butters");
plt.show();
```

```python
# Historgrams
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
plt.show();
```

# Two options for subplots

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2);
#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), height=nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
plt.show();
```

```python
# Subplots Option 2
fig, ax = plt.subplots(nrows=2, ncols=2);
#plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), height=nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
plt.show();
```

### Which one should you use?
### pyplot vs matplotlib OO method
 * When plotting something quickly, its okay to use the pyplot method
 * When plotting something more advanced, use the OO method


```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
# This is pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10,6));
plt.show();
```

```python
## this is OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim([45,100]);
plt.show();
```

```python
## this is OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']);

# Customize the plot
ax.set(title="Heart Disaese and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Targets");

#Add a horizontol line
ax.axhline(over_50["chol"].mean(), linestyle='--');
plt.show();
```

```python
# Subplot of chol, age, talach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True);

#Add data to ax0
# Plot the data
scatter = ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']);

# Customize the plot
ax0.set(title="Heart Disaese and Cholestrol Levels",  ylabel="Cholestrol");

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Targets");

#Add a horizontol line
ax0.axhline(over_50["chol"].mean(), linestyle='--');

#Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                    c=over_50['target']);

# Customize the plot
ax1.set(title="Heart Disaese and Max Heart Rate", xlabel="Age", ylabel="Heart Beats level");

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Targets");

#Add a horizontol line
ax1.axhline(over_50["thalach"].mean(), linestyle='--');
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

plt.show();
```

```python
plt.style.use('seaborn-v0_8-whitegrid');
## this is OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter'); # set the color scheme

# Customize the plot
ax.set(title="Heart Disaese and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Targets");

#Add a horizontol line
ax.axhline(over_50["chol"].mean(), linestyle='--');
plt.show();
```

```python
plt.style.use('seaborn-v0_8-whitegrid');
# Subplot of chol, age, talach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True);

#Add data to ax0
# Plot the data
scatter = ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter'); # set the color schem;

# Customize the plot
ax0.set(title="Heart Disaese and Cholestrol Levels",  ylabel="Cholestrol");

# Set the x axis limit
ax0.set_xlim([50, 80]);

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Targets");

#Add a horizontol line
ax0.axhline(over_50["chol"].mean(), linestyle='--');

#Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                    c=over_50['target'],
                    cmap='winter'); # set the color scheme

# Customize the plot
ax1.set(title="Heart Disaese and Max Heart Rate", xlabel="Age", ylabel="Heart Beats level");

# Set the x axis limit
ax1.set_xlim([50, 80]);

# Set the y axis limit
ax1.set_ylim([60, 200]);

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Targets");

#Add a horizontol line
ax1.axhline(over_50["thalach"].mean(), linestyle='--');
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

plt.show();
```
